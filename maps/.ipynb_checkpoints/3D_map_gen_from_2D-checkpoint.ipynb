```python
# Reads the data of a single experiment
def read_mapfile(name):
    fname = MAP_DIR + name + ".map"
    print(fname)
    # Read data
    with open(fname) as csvfile:
        data = [row for row in csv.reader(csvfile,delimiter=',' )]
    return pd.DataFrame(data = data)
```

```python
Map_Height = 34 # X coordinate
Map_Length = 29 # Y coordinate
OUT_DIR = "/home/vivek/Vivek/Projects/Connectivity/ARGOS_Planning/maps_custom/kh4"
OUT_FNAME = 'obsE.map'
ARENA ={'MIN_X':0.18,'MAX_X':3.6,'MIN_Y':-2.22, 'MAX_Y':0.75}
WALL_THICKNESS = 0.3
MAP_SCALE = 10
ARENA_WALL ={}
ARENA_WALL['MIN_X'] = ARENA['MIN_X']+WALL_THICKNESS
ARENA_WALL['MAX_X'] = ARENA['MAX_X']-WALL_THICKNESS
ARENA_WALL['MIN_Y'] = ARENA['MIN_Y']+WALL_THICKNESS
ARENA_WALL['MAX_Y'] = ARENA['MAX_Y']-WALL_THICKNESS
ARENA_WALL['RANGE_X']=ARENA_WALL['MAX_X']-ARENA_WALL['MIN_X']
ARENA_WALL['RANGE_Y']=ARENA_WALL['MAX_Y']-ARENA_WALL['MIN_Y']

ARENA['RANGE_X']=ARENA['MAX_X']-ARENA['MIN_X']
ARENA['RANGE_Y']=ARENA['MAX_Y']-ARENA['MIN_Y']
obs = []
obs1 ={'x_min':1.1,'x_max':1.5,'y_min':-1.2,'y_max':0.5}
obs.append(obs1)
obs2 ={'x_min':2.3,'x_max':2.9,'y_min':-2.0,'y_max':-0.23}
obs.append(obs2)
# obs2 ={'x_min':0.18,'x_max':3.6,'y_min':-2.22,'y_max':0.75}
# obs.append(obs2)
```

```python
def obtain_tree_str_for_len(len_of_str):
    r_str=''
    for s in range(0,len_of_str):
        r_str = r_str+'T'
    return r_str
def create_str_line(border,length):
    if(border == 1):
        r_str='T'
        for i in range(1,length-1):
            r_str = r_str+'.'
        r_str=r_str+'T'
    return r_str
def transorm_obs_to_map_coor(obst,arena):
    for ob in obst:
        ob['x_min'] = (Map_Height * (ob['x_min']-arena['MIN_X']))/arena['RANGE_X']
        ob['x_max'] = (Map_Height * (ob['x_max']-arena['MIN_X']))/arena['RANGE_X']
        ob['y_min'] = (Map_Length * (ob['y_min']-arena['MIN_Y']))/arena['RANGE_Y']
        ob['y_max'] = (Map_Length * (ob['y_max']-arena['MIN_Y']))/arena['RANGE_Y']
        
def transform_obs_to_arena_coor(obst,arena):
    for ob in obst:
        ob['x_min'] = ( (arena['RANGE_X'] * ob['x_min'])/Map_Height) + arena['MIN_X']
        ob['x_max'] = ( (arena['RANGE_X'] * ob['x_max'])/Map_Height) + arena['MIN_X']
        ob['y_min'] = ( (arena['RANGE_Y'] * ob['y_min'])/Map_Length) + arena['MIN_Y']
        ob['y_max'] = ( (arena['RANGE_Y'] * ob['y_max'])/Map_Length) + arena['MIN_Y']
def invert_y_min_max(obst,arena):
    for ob in obst:
        ob['y_min'] = Map_Length - ob['y_min']
        ob['y_max'] = Map_Length - ob['y_max']
        
def obtain_T_str_for_len(len_of_str):
    r_str=''
    for s in range(0,len_of_str):
        r_str = r_str+'T'
    return r_str
def obtain_dot_str_for_len(len_of_str):
    r_str=''
    for s in range(0,len_of_str):
        r_str = r_str+'.'
    return r_str
```

```python
khepare_2d_map =pd.DataFrame()
df_original.applymap(str)
# write index of df
khepare_2d_map=khepare_2d_map.append(['type octile'],ignore_index = True)
khepare_2d_map=khepare_2d_map.append(['height '+str(Map_Height)],ignore_index = True)
khepare_2d_map=khepare_2d_map.append(['width '+str(Map_Length)],ignore_index = True)
khepare_2d_map=khepare_2d_map.append(['map'],ignore_index = True)

#TOP wall
for i in range(0,int(WALL_THICKNESS*MAP_SCALE) ):
    khepare_2d_map = khepare_2d_map.append([obtain_tree_str_for_len(Map_Length)],ignore_index = True)

#MAP rows This assumes no two obs collides 
for i in range(int((WALL_THICKNESS*MAP_SCALE)), int(Map_Height-(WALL_THICKNESS*MAP_SCALE))):
    row_min = 0
    row_max = Map_Length
    row_str = obtain_T_str_for_len(int(WALL_THICKNESS*MAP_SCALE)) # SIDE WALL
    for ob in obs:
        if(i> int(ob['x_min']) and i< int(ob['x_max'])):
            row_min = ob['y_max']
            row_max = ob['y_min']   
    if(row_min > 0):
        before_obs = 0
        if(row_min > int(WALL_THICKNESS*MAP_SCALE) and int(row_min - int(WALL_THICKNESS*MAP_SCALE)) > 0):
            before_obs= int(row_min) - int(WALL_THICKNESS*MAP_SCALE)
            print before_obs
            row_str = row_str+obtain_dot_str_for_len(int(before_obs))
             
        row_str = row_str+obtain_T_str_for_len(int(row_max-row_min))
        after_obs = 0
        if(row_max < Map_Length and int(Map_Length- row_max)>0):
            after_obs= int(Map_Length- row_max)
            row_str = row_str+obtain_dot_str_for_len(int(after_obs))
        row_str = row_str+obtain_T_str_for_len(int(WALL_THICKNESS * MAP_SCALE))
        
    else:
        row_str = row_str+obtain_dot_str_for_len(int(Map_Length-((WALL_THICKNESS*MAP_SCALE)*2))) 
        row_str = row_str+ obtain_T_str_for_len(int(WALL_THICKNESS*MAP_SCALE))
    khepare_2d_map = khepare_2d_map.append([row_str],ignore_index = True)
    
khepare_2d_map = khepare_2d_map.append([obtain_tree_str_for_len(Map_Length)],ignore_index = True)
print khepare_2d_map
```
